# Contexto del Dataset y Caso de Estudio

El presente dataset proviene de una cooperativa financiera y contiene información consolidada sobre los créditos otorgados a sus asociados. Cada registro representa un préstamo activo o histórico, e integra variables de tres grandes dimensiones: las condiciones financieras del crédito (monto, plazo, cuotas, saldos de capital e intereses), el perfil socioeconómico del cliente (edad, sexo, estrato, ingresos, egresos y actividad económica) y su relación con la cooperativa (tiempo de vinculación, aportes, cuentas de ahorro y garantías). Adicionalmente, el dataset incluye variables geográficas que permiten identificar la ubicación del asociado a nivel de departamento y ciudad, así como variables agrupadas que facilitan el análisis por segmentos.

El objetivo central de este análisis es predecir la variable `default`, la cual indica si un cliente ha incurrido en impago, entendido no como un simple retraso en el pago de sus cuotas, sino como el cese definitivo del cumplimiento de sus obligaciones crediticias. Esta variable es de naturaleza binaria, donde **1** representa que el cliente entró en default y **0** que se encuentra al día con sus pagos. En este contexto, el análisis exploratorio de datos (EDA) busca identificar patrones, distribuciones y relaciones entre las variables que permitan comprender qué factores están más asociados al incumplimiento, sentando así las bases para la construcción de un modelo predictivo de riesgo crediticio que apoye la toma de decisiones de la cooperativa en el otorgamiento y seguimiento de créditos.
